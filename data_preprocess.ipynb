# package import and function define

```python
import pandas as pd
import pickle as pkl
import numpy as np
import os
import glob
```

```python
def pkl_dump(data, file):
    with open(file, "wb") as fw:
        pkl.dump(data, fw)

        
def pkl_load(file):
    with open(file, "rb") as fr:
        data = pkl.load(fr)
    return data
```

```python
data_root='/data/userData/zehao.yu/drug_repropose/'
```

```python
print(glob.glob(data_root+"*.csv"))
```

```python
def agg_table_by_pid(df, d):
    for pid, sub_df in df.groupby('Subject_ID'):
        del sub_df['Subject_ID']
        if pid in d.keys():
            d[pid]=[*d[pid],*[x for _, x in sub_df.to_dict('index').items()]]
        else:
            d[pid] = [x for _, x in sub_df.to_dict('index').items()]
```

```python
df_iter1 = pd.read_csv(data_root+'procedures.csv',chunksize=100)
for iter_num, chunk in enumerate(df_iter1, 1):
    print(chunk)
    break
```

# demographic

```python
d=dict()
df_iter1 = pd.read_csv(data_root + 'demographic.csv',dtype=str, chunksize=5*10**6, iterator=True)
# chunk loop
for iter_num, chunk in enumerate(df_iter1, 1):
    print(f'diagnosis Processing iteration {iter_num}')
    dat1=chunk
    agg_table_by_pid(dat1, d)
```

# procedures

```python
d2=dict()
df_iter1 = pd.read_csv(data_root + 'procedures.csv',dtype=str, chunksize=100, iterator=True)
# chunk loop
for iter_num, chunk in enumerate(df_iter1, 1):
    #print(f'diagnosis Processing iteration {iter_num}')
    dat1=chunk[['Subject_ID','PX_DATE','PX','PX_TYPE']]
    #print(dat1)
    #break
    agg_table_by_pid(dat1, d2)
```

```python
for k, v in d2.items():
    d[k][0]['procedures'] = [v]
```
